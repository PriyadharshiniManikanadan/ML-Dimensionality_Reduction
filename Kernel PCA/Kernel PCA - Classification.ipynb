# Kernel PCA - Classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import KernelPCA
```

```python
# Data Collection

raw_data = pd.read_csv("CKD.csv")
raw_data
```

```python
df = raw_data
```

```python
df = pd.get_dummies(df,dtype = int, drop_first=True)
```

```python
# Input Output Split

indep_x = df.drop('classification_yes',axis = 1)
dep_y = df['classification_yes']
```

```python
def split_scaler(indep_x,dep_y):
    x_train,x_test,y_train,y_test = train_test_split(indep_x,dep_y,test_size = 0.2, random_state=0)
    sc=StandardScaler()
    x_train = sc.fit_transform(x_train)
    x_test = sc.transform(x_test)
    return x_train,x_test,y_train,y_test

def cm_prediction(classifier,x_test,y_test):
    y_pred = classifier.predict(x_test)
    cm = confusion_matrix(y_test,y_pred)
    Accuracy = accuracy_score(y_test,y_pred)
    Report = classification_report(y_test,y_pred)
    return cm,Accuracy,Report

def logistic(x_train,y_train,x_test,y_test):
    classifier = LogisticRegression(random_state=0)
    classifier.fit(x_train,y_train)
    return cm_prediction(classifier,x_test,y_test)

def svm_linear(x_train,y_train,x_test,y_test):
    classifier = SVC(kernel='linear',random_state=0)
    classifier.fit(x_train,y_train)
    return cm_prediction(classifier,x_test,y_test)

def svm_nonlinear(x_train,y_train,x_test,y_test):
    classifier = SVC(kernel = 'rbf', random_state=0)
    classifier.fit(x_train,y_train)
    return cm_prediction(classifier,x_test,y_test)

def knn(x_train,y_train,x_test,y_test):
    classifier = KNeighborsClassifier(n_neighbors=5, metric = 'minkowski')
    classifier.fit(x_train,y_train)
    return cm_prediction(classifier,x_test,y_test)

def naive(x_train,y_train,x_test,y_test):   
    classifier = GaussianNB()
    classifier.fit(x_train,y_train)
    return cm_prediction(classifier,x_test,y_test)

def decision(x_train,y_train,x_test,y_test):
    classifier = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
    classifier.fit(x_train,y_train)
    return cm_prediction(classifier,x_test,y_test)

def random(x_train,y_train,x_test,y_test):
    classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state=0)
    classifier.fit(x_train,y_train)
    return cm_prediction(classifier,x_test,y_test)

def KernelPCA_results(acclog,accsvml,accsvmnl,accknn,accnav,accdc,accrf):
    KernelPCAdataframe = pd.DataFrame(index = ['Logistic', 'SVM', 'Decision', 'Random'], columns = ['Logistic', 'SVM_l', 'SVM_nl', 'KNN',
                                                                                              'Naive','Decision','Random'])
    for index in KernelPCAdataframe.index:
        KernelPCAdataframe['Logistic'][index] = acclog
        KernelPCAdataframe['SVM_l'][index] = accsvml
        KernelPCAdataframe['SVM_nl'][index] = accsvmnl
        KernelPCAdataframe['KNN'][index] = accknn
        KernelPCAdataframe['Naive'][index] = accnav
        KernelPCAdataframe['Decision'][index] = accdc
        KernelPCAdataframe['Random'][index] = accrf
    return KernelPCAdataframe
```

```python
acclog = []
accsvml = []
accsvmnl = []
accknn = []
accnav = []
accdc = []
accrf = []
```

```python
# Train Test Split

x_train, x_test, y_train, y_test = split_scaler(indep_x,dep_y)
```

```python
# Kernel PCA Model Creation

KPCA = KernelPCA(n_components = 1, kernel ='linear')
x_train_kpca = KPCA.fit_transform(x_train)
x_test_kpca = KPCA.transform(x_test)
```

```python
# Calculate accuracies for each classifier

cm, Accuracy, Report = logistic(x_train_kpca,y_train,x_test_kpca,y_test)
acclog.append(Accuracy)

cm, Accuracy, Report = svm_linear(x_train_kpca,y_train,x_test_kpca,y_test)
accsvml.append(Accuracy)

cm, Accuracy, Report = svm_nonlinear(x_train_kpca,y_train,x_test_kpca,y_test)
accsvmnl.append(Accuracy)

cm, Accuracy, Report = knn(x_train_kpca,y_train,x_test_kpca,y_test)
accknn.append(Accuracy)

cm, Accuracy, Report = naive(x_train_kpca,y_train,x_test_kpca,y_test)
accnav.append(Accuracy)

cm, Accuracy, Report = decision(x_train_kpca,y_train,x_test_kpca,y_test)
accdc.append(Accuracy)

cm, Accuracy, Report = random(x_train_kpca,y_train,x_test_kpca,y_test)
accrf.append(Accuracy)

result = KernelPCA_results(acclog[0],accsvml[0],accsvmnl[0],accknn[0],accnav[0],accdc[0],accrf[0])
```
